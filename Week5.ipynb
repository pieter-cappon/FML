```python
import seaborn as sns
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
```

```python
df = pd.read_csv('Employee-Attrition.csv')
df.head()
```

```python
df['Attrition'].value_counts()
```

```python
# first off we need dummy variables to be ablo to predict Attrition, it is now Yes and No. This has to be 0 and 1. 
# i use concat to add the dummy values to the df
dummies = pd.get_dummies(df['Attrition'])
df = pd.concat([df, dummies], axis=1) 
df.head(1)
```

```python
# now that attrition has dummies I can look at it in the correlation matrix
df.corr()
```

```python
# based on the correlations I picked the following variables
df_subset = df[['Yes','YearsInCurrentRole', 'TotalWorkingYears', 'JobLevel', 'Age', 'MonthlyIncome', 'YearsWithCurrManager', 'YearsAtCompany']]
df_subset.head()
```

<h2> K- nearest neighbour</h2>


```python
from sklearn.preprocessing import normalize #get the function needed to normalize our data.

X = df_subset[['YearsInCurrentRole', 'TotalWorkingYears', 'JobLevel', 'Age', 'MonthlyIncome', 'YearsWithCurrManager', 'YearsAtCompany']] #create the X matrix
X = normalize(X) #normalize the matrix to put everything on the same scale
y = df_subset['Yes'] #create the y-variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #split the data, store it into different variables
```

```python
from sklearn.neighbors import KNeighborsClassifier #the object class we need

knn = KNeighborsClassifier(n_neighbors=5) #create a KNN-classifier with 5 neighbors (default)
knn = knn.fit(X_train, y_train) #this fits the k-nearest neigbor model with the train data
knn.score(X_test, y_test) #calculate the fit on the test data
```

```python
from sklearn.metrics import confusion_matrix
y_test_pred = knn.predict(X_test) #the predicted values
cm = confusion_matrix(y_test, y_test_pred) #creates a "confusion matrix"
cm
```

```python
#In order to read it, I made a confusion matrix in a table and added the labels
conf_matrix = pd.DataFrame(cm, index=['No Attrition', 'Attrition'], columns = ['No Attrition_p', 'Attrition_p']) 
conf_matrix
```

```python
# calculating f1-score, recall and precision 

from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_pred))
```
