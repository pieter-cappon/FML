```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("class - Blad1.csv")
df.head()
```

```python
df['linkedin_con'].name = 'LinkedIn' #Labeling the variable for plotting
linked_in = df['linkedin_con'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(linked_in) 
plt.title('Connections on LinkedIn')
plt.show()
```

```python
sns.boxplot(df['phones_owned'])
plt.title('Amount of phones owned')
plt.xlabel('Phones')
plt.show()
```

```python
sns.violinplot(df['instagram_followers'])
sns.swarmplot(df['instagram_followers'], color='purple') #default color doesn't stand out
plt.title('Follower count: Instagram')
plt.ylabel('Frequency')
plt.xlabel('Followers')
plt.show()
```

```python
df['apps_phone'].name = 'Apps on phone' #Labeling the variable for plotting
apps = df['apps_phone'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(apps, kde=False) 
plt.title('Connections on LinkedIn')
plt.show()
```
